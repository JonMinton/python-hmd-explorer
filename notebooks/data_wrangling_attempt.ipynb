```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import plot

dta = pd.read_csv('../assets/data/lifetables.csv')
```

```python
filtered_dta = dta[(dta['sex'] == 'male') & (dta['cntry'] == 'UKR') & (dta['age'] == 0)][['year', 'ex']]
```

```python
filtered_dta.head()
```

```python
# filtered_dta.plot(x='year', y = 'ex', kind ='line')

trace = go.Line(
    x=filtered_dta['year'],
    y=filtered_dta['ex'],
    mode='lines'
)

layout = go.Layout(
    title = "Plot",
    xaxis = dict(title = 'Year'),
    yaxis = dict(title = 'Life expectancy at birth')
)

fig = go.Figure(data = [trace], layout = layout)

plot(fig)
```
