```python
print(eligibleCountries)

import rpy2.robjects as robjects 
from rpy2.robjects import pandas2ri

pythonEligibleCountries = pd.DataFrame(eligibleCountries).transpose()

pythonEligibleCountries.columns = ['country', 'link', 'code']
```

Now, for each code, I want to call item availability



```python
allHmdItems = []

for countryCode in pythonEligibleCountries.code:
    thisCountryInfo = {'code': countryCode, 'itemInfo': hmdHfdPlus.getHMDitemavail(countryCode)}
    allHmdItems.append(thisCountryInfo)


```
